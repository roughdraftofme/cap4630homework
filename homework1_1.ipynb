```
## declarations
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
import numpy as np
from mpl_toolkits import mplot3d

def printContour(X, Y, Z):
  plt.contour(X, Y, Z, 30, cmap='BuGn')
  plt.colorbar()
  
def printSurface(X, Y, Z, point):
  fig = plt.figure()
  ax = plt.axes(projection='3d')
  ax.contour3D(X, Y, Z, 50, cmap='binary')
  ax.scatter3D(0, 0, point, marker='o', cmap='Greens');
  ax.set_xlabel('x')
  ax.set_ylabel('y')
  ax.set_zlabel('z')
  ax.view_init(55)
```

```
## part one

def f1(x, y):
  return x ** 2 + y ** 2

# for the contour plot
x = np.linspace(-10, 10, 30)
y = np.linspace(-10, 10, 30)

X, Y = np.meshgrid(x, y)
Z = f1(X, Y)

printContour(X, Y, Z)

# now using the same values on a surface plot
printSurface(X, Y, Z, f1(0,0))
```

```
## part two

def f2(x, y):
  return -(x ** 2 + y ** 2)


# for the contour plot
x = np.linspace(-10, 10, 30)
y = np.linspace(-10, 10, 30)

X, Y = np.meshgrid(x, y)
Z = f2(X, Y)

printContour(X, Y, Z)

# now using the same values on a surface plot
printSurface(X, Y, Z, f2(0,0))
```

```
## part three

def f3(x, y):
  return x ** 2 - y ** 2


# for the contour plot
x = np.linspace(-10, 10, 30)
y = np.linspace(-10, 10, 30)

X, Y = np.meshgrid(x, y)
Z = f3(X, Y)

printContour(X, Y, Z)

# now using the same values on a surface plot
printSurface(X, Y, Z, f3(0,0))
```
